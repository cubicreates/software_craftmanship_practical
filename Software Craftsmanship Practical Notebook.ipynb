```python
class Calculator:
    def __init__(self):
        self.history = {}  # Dictionary to keep track of calculation history

    def add(self, a, b):
        result = a + b
        ex = f"{a} + {b}"
        self.history[ex]=result
        return result

    def subtract(self, a, b):
        result = a - b
        ex = f"{a} - {b}"
        self.history[ex]=result
        return result

    def multiply(self, a, b):
        result = a * b
        ex = f"{a} x {b}"
        self.history[ex]=result
        return result

    def divide(self, a, b):
        if b == 0:
            ex = f"{a} / {b}"
            self.history[ex]='Division By Zero'
            return "Error: Division by zero"
        else:
            result = a / b
            ex = f"{a} / {b}"
            self.history[ex]=result
            return result

    def modulus(self, a, b):
        if b == 0:
            ex = f"{a} + {b}"
            self.history[ex]='Division By Zero'
            return "Error: Division by zero"
        else:
            result = a % b
            ex = f"{a} + {b}"
            self.history[ex]=result
            return result

    def show_history(self):
        if not self.history:
            return "No calculations yet."
        else:
            for i in self.history:
                return f"{i} = {self.history[i]}"

def main():
    calc = Calculator()

    print("\t Calculator \n ~Adit Ghosh")
    while True:
        print("\nOptions:")
        print("Type '1' for Addition")
        print("Type '2' for Subtraction")
        print("Type '3' for Multiplication")
        print("Type '4' for Division")
        print("Type '5' for Modulus")
        print("Type '6' for Show History")
        print("Type '7' for Exit")

        choice = int(input("Select an option (1-7): "))

        if choice == 7:
            print("Thank You for Using.")
            break
        elif choice == 6:
            print("\nCalculation History:")
            print(calc.show_history())
        elif choice in {1, 2, 3, 4, 5}:
            try:
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))

                if choice == 1:
                    print(">>>", calc.add(num1, num2))
                elif choice == 2:
                    print(">>>", calc.subtract(num1, num2))
                elif choice == 3:
                    print(">>>", calc.multiply(num1, num2))
                elif choice == 4:
                    print(">>>", calc.divide(num1, num2))
                elif choice == 5:
                    print(">>>", calc.modulus(num1, num2))
            except ValueError:
                print("Invalid input. Please enter integer or decimal values.")
        else:
            print("Invalid choice. Please select a valid option.")

main()
```
